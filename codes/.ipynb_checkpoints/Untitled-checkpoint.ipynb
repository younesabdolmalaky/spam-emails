```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('../datasets/Corona_NLP_train.csv')
df = df[['TweetAt' , 'OriginalTweet' , 'Sentiment']]
df['TweetAt'] = pd.to_datetime(df['TweetAt'], format='%d-%m-%Y' ,  errors='coerce')
df
```

```python
df['WeekDay'] = df['TweetAt'].dt.day_name()
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0)
```

```python
df = pd.get_dummies(df, columns = ['WeekDay'])
```

```python
df = df.drop(['TweetAt'], axis=1)
```
